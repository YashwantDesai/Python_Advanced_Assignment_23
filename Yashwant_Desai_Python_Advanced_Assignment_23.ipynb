# Q1. If you have any, what are your choices for increasing the comparison between different figures on the same graph?

To increase the comparison between different figures on the same graph in Python, consider the following key techniques:

Color and Style:Use different colors, line styles, and markers for each figure.
Legends:Add a legend to label each figure.
Gridlines:Include gridlines to provide a common reference.
Plot Titles and Labels:Clearly label the figures with titles and axis labels.
Transparency:Adjust the transparency (alpha) of figures to overlay and compare.
Secondary Y-Axis:Use a secondary y-axis if your data has different units or scales.

Data Range:Adjust the axis range to focus on specific data regions.

By combining these techniques and using appropriate plotting libraries like Matplotlib or Seaborn, you can effectively enhance the comparison of different figures on a single graph in Python.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 1, 3, 5]
y2 = [1, 3, 2, 4, 1]

plt.plot(x, y1, color='blue', linestyle='-', marker='o', label='Figure 1')

plt.plot(x, y2, color='red', linestyle='--', marker='x', label='Figure 2')

plt.legend()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Comparison of Two Figures')

plt.grid()

plt.show()
```

# Q2. Can you explain the benefit of compound interest over a higher rate of interest that does not compound after reading this chapter?

Compound interest is the interest calculated not only on the initial principal but also on any interest earned in previous periods. This means your money grows at an accelerating rate. In contrast, simple interest is calculated only on the initial principal amount and remains constant throughout the investment or loan term. Below is an example.

```python
def calculate_simple_interest(principal, rate, time):
    return principal * (1 + (rate * time))
def calculate_compound_interest(principal, rate, time, n):
    return principal * (1 + (rate / n))**(n * time)

principal = 100000  
rate = 0.06       
time = 5          
n = 12            

simple_interest = calculate_simple_interest(principal, rate, time)
compound_interest = calculate_compound_interest(principal, rate, time, n)

print(f"Simple Interest after {time} years: {simple_interest:.2f} Rs")
print(f"Compound Interest after {time} years: {compound_interest:.2f} Rs")

interest_difference = compound_interest - simple_interest
print(f"Difference between Compound and Simple Interest: {interest_difference:.2f} Rs")
```

# Q3. What is a histogram, exactly? Name a numpy method for creating such a graph.

A histogram is a graphical representation of the distribution of a dataset. It divides the data into bins or intervals and displays the frequency or count of data points within each bin as vertical bars. In Python, you can create a histogram using the numpy.histogram() function, which computes the bin edges and counts. You can then use a plotting library like Matplotlib to visualize the histogram by specifying the data, bin edges, and other plot attributes. Histograms are valuable for gaining insights into the distribution of data, identifying patterns, and understanding the central tendencies and variations within a dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000) 

hist, bin_edges = np.histogram(data, bins=10)

plt.hist(data, bins=bin_edges, edgecolor='k')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# Q4. If necessary, how do you change the aspect ratios between the X and Y axes?

We can change the aspect ratios between the X and Y axes in a Python plot, specifically when using the matplotlib library, by using the set_aspect function with different options:

'equal': This option ensures an equal aspect ratio, meaning one unit on the X-axis is the same physical length as one unit on the Y-axis.

'auto': The aspect ratio is determined automatically based on the data and plot dimensions.

A numerical value: You can set a specific aspect ratio by providing a numerical value, e.g., set_aspect(2), to customize the scaling of the X and Y axes.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [15, 8, 6, 4, 2]

plt.plot(x, y)
plt.gca().set_aspect('equal')
plt.title('Equal Aspect Ratio')

plt.figure()
plt.plot(x, y)
plt.gca().set_aspect('auto')
plt.title('Auto Aspect Ratio')

plt.figure()
plt.plot(x, y)
plt.gca().set_aspect(2)
plt.title('Custom Aspect Ratio (2:1)')

plt.show()
```

# Q6. Before you buy a home, which numpy function will you use to measure your monthly mortgage payment?

Numpy does not have a built-in function for calculating mortgage payments directly. Mortgage payment calculations typically require financial functions that are not available in the NumPy library.
We can consider using specialized financial libraries, such as numpy_financial

```python
import numpy_financial as npf

principal = 200000  
annual_interest_rate = 8  
loan_term_years = 30  

monthly_interest_rate = annual_interest_rate / 12 / 100

num_payments = loan_term_years * 12

monthly_payment = npf.pmt(rate=monthly_interest_rate, nper=num_payments, pv=-principal)

print(f"Your monthly mortgage payment will be: {monthly_payment:.2f} Rs")
```

# Q7. Can string data be stored in numpy arrays? If so, list at least one restriction that applies to this data.

Yes, string data can be stored in NumPy arrays. NumPy provides a data type called numpy.str_ to handle string data. We can create NumPy arrays of string data by specifying the data type as numpy.str_, and you can store strings of various lengths in these arrays.

There is one significant restriction when working with NumPy arrays that contain strings: fixed-length strings. In a NumPy string array, the strings are of a fixed length, and any string stored in the array must have the same length as specified when creating the array. This can lead to inefficient memory usage when storing variable-length strings, as all strings will be padded to the fixed length.

```python
import numpy as np

string_array = np.array(['yash', 'umesh', 'ramesh'], dtype=np.str_)

string_array[0] = 'yashwant'  
string_array[1] = 'desai'   

print(string_array)
```
